```python
# import the required libraries
import os
import re
import csv
import pandas as pd
import numpy as np
```

```python
# Set variables for the project (i.e. the input location of the file to be processed and the output location) )

folderlink = 'data//bioportal//'
input_file_bio = os.path.join(folderlink, '2019_12_10_BioPort_BPR_MASTERFILE.csv')
input_file_bio_name_var = os.path.join(folderlink, '2019_12_10_BioPort_BPR_name_var.csv')
input_file_bio_prov = os.path.join(folderlink, '2019_12_10_BioPort_BPR_provenance.csv')
folder_output = 'data//output//'
input_file_parent = folder_output+"parent_data.csv"
```

```python
# Panda settings for showing data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
bio = pd.read_csv(input_file_bio, sep=';', encoding='utf-8')
```

```python
bio['birth_year'] = bio['event_birth_when'].str.extract(r'(\d{3})')
```

```python
bio['first_letter'] = bio['voornaam'].astype(str).apply(lambda x: x[0] + '.')
```

```python
bio['bio_name_surname_year'] = (bio['first_letter'].astype(str) + '_' + np.where(bio['intrapositie'].isna(), '', bio['intrapositie'].astype(str)) + '_' + bio['geslachtsnaam'].astype(str) + '_' + np.where(bio['birth_year'].isna(), '', bio['birth_year'].astype(str)))
```

```python
year_int = {'year_birth': pd.Int64Dtype()}
```

```python
parent = pd.read_csv(input_file_parent, sep=';', dtype=year_int , encoding='utf-8')
```

```python
parent['first_letter'] = parent['name'].astype(str).apply(lambda x: x[1] + '.')

```

```python
parent['year_birth3'] = parent['year_birth'].astype(str)
```

```python
parent['year_birth2']= parent['year_birth3'].str.extract(r'(\d{3})')
```

```python
parent.head()
```

```python
parent['par_name_surname_year'] = (parent['first_letter'].astype(str) + '_' + np.where(parent['infix'].isna(), '', parent['infix'].astype(str)) + '_' + parent['surname'].astype(str) + '_' + np.where(parent['year_birth2'].isna(), '', parent['year_birth2'].astype(str)))
```

```python
parent.head()
```

```python
parent_bio = pd.merge(parent, bio, left_on='par_name_surname_year', right_on='bio_name_surname_year', how='inner')
```

```python
parent_bio.tail()
```

```python
parent_bio.to_csv(folderlink+'parent_bio_3y.csv', sep=';', encoding='utf-8', index=False)
```
