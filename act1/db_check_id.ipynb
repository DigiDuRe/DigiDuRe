```python
# Set variables for the project (i.e. the input location of the file to be processed and the output location) )

folderlink = 'output//'
```

```python
d1_csv = folderlink+"parent_data.csv.csv" 
d11_csv = folderlink+"minister_info.csv"
d22_csv = folderlink+"d22.csv"
```

```python
d1 = pd.read_csv(d1_csv, sep=';', encoding='utf-8')
```

```python
d11_col_data_type = {'minister_year': pd.Int64Dtype()}
```

```python
d11 = pd.read_csv(d11_csv, sep=';', dtype=d11_col_data_type, encoding='utf-8')
```

```python
d22 = pd.read_csv(d22_csv)
```

```python
d22.head()
```

```python
d11.head()
```

```python
d22['d22join_name'] = d22['predikant'].str.replace(";","").str.replace(" ","").str.replace("  ","")
d22['d22gemeente'] = d22['gemeente'].str.replace(";","").str.replace(" ","").str.replace("  ","")
d22['d22jaar intrede'] = d22['jaar intrede'].astype(str).str.replace(";","").str.replace("","").str.replace("  ","")
```

```python
d11['d11join_name'] = d11['join_name'].str.replace(";","").str.replace(" ","").str.replace("  ","")
d11['d11minister']= d11['minister'].str.replace(";","").str.replace(" ","").str.replace("  ","")
d11['d11minister_year']= d11['minister_year'].astype(str).str.replace(";","").str.replace("","").str.replace("  ","")
d11['d11id'] = d11['id']
```

```python
d22.head()
```

```python
d11['d11j_place_year']=d11['d11join_name'].astype(str)+"_"+d11['d11minister'].astype(str)+"_"+d11['d11minister_year'].astype(str)
```

```python
d11.head()
```

```python
d22['d22j_place_year']=d22['d22join_name'].astype(str)+"_"+d22['d22gemeente'].astype(str)+"_"+d22['d22jaar intrede'].astype(str)
```

```python
d11_d22 = pd.merge(d11, d22, left_on='d11j_place_year', right_on='d22j_place_year', how='inner')
```

```python
id_counts_d11 = d11['id'].value_counts()
```

```python
d11['d11_count'] = d11['id'].map(id_counts_d11)
```

```python
d11['d11_count_unique']= d11['d11id'].astype(str)+"__"+d11['d11_count'].astype(str)
```

```python
d11.head()
```

```python
id_counts_d22_d11 = d11_d22['d11id'].value_counts()
d11_d22['d11_d22_count'] = d11_d22['d11id'].map(id_counts_d22_d11)
```

```python
d11_d22['d11_d22_count_unique']= d11_d22['d11id'].astype(str)+"__"+d11_d22['d11_d22_count'].astype(str)
```

```python
countd11_countd11_d22 = pd.merge(d11, d11_d22, left_on='d11_count_unique', right_on='d11_d22_count_unique', how='inner')
```

```python
countd11_countd11_d22.head()
```

```python
output_table = countd11_countd11_d22[['id_x','d11_count_unique', 'd11_d22_count_unique']].copy()
```

```python
output_table.head()
```

```python
output_without_duplicates = output_table.drop_duplicates()
```
