```python
import docx2txt

# Set the path to your input and output files
input_file = 'C://temp//Repertoriummetoudepersoonsnummers.docx'
temp_file = 'C://temp//temp.txt'

# Use docx2txt library to extract text from .docx file
text = docx2txt.process(input_file)

# Write the extracted text to a .txt file
with open(temp_file, 'w', encoding='utf-8') as f:
    f.write(text)
```

```python
output_file = 'C://temp//output.txt'
```

```python
with open(temp_file, 'r', encoding='utf-8') as file:
    data = file.read()

# Replace newlines with semicolons
data = data.replace('\n\n', ';')

# Remove excessive whitespaces
data = ' '.join(data.split())

# Replace semicolons with newlines and add semicolons between name and details
data = data.replace('; ', ';\n').replace(';','; ')
```

```python
output_file_2 = 'C://temp//output2.txt'
with open(output_file_2, 'w', encoding='utf-8') as f:
    f.write(data)
```

```python
output_file_3 = 'C://temp//output3.txt'
output_file_4 = 'C://temp//output4.txt'
```

```python
with open(output_file_2, "r") as input_file:
    with open(output_file_3, "w") as output_file:
        merged_line = ''
        for line in input_file:
            if '<'and'>' in line:
                # if <x> is present, write the merged line to output file and reset the merged_line
                output_file.write(merged_line + line.strip() + '\n')
                merged_line = ''
            else:
                # otherwise merge the current line to the merged_line
                merged_line += line.strip() + ' '
```

```python
with open(output_file_3, "r") as f_in:
    with open(output_file_4, "w") as f_out:
        for line in f_in:
            if line.startswith("; "):
                line = line[2:]
            f_out.write(line)
```

```python
import csv

input_file = output_file_4
output_file = 'C://temp//output_file.csv'

# Define the headers for the output file
headers = ['Surname', 'Name', 'Field 3', 'Field 4', 'Field 5', 'Field 6', 'Field 7', 'Field 8', 'Field 9', 'Field 10', 'Field 11', 'Field 12', 'Field 13']

with open(input_file, 'r') as infile, open(output_file, 'w', newline='') as outfile:
    reader = csv.reader(infile, delimiter=';')
    writer = csv.writer(outfile)
    
    # Write the headers to the output file
    writer.writerow(headers)

    # Loop through each row in the input file and write it to the output file with 12 fields
    for row in reader:
        # Create a new row with 12 fields by extending the current row with empty values
        new_row = row + [''] * (12 - len(row))
        writer.writerow(new_row)
```
