```python
import pandas as pd
import os
from SPARQLWrapper import SPARQLWrapper, JSON
```

```python
folderlink = '..//data//'
input_folder = 'input//'
folder_output = 'output//'
ddrc_file = 'parent_data.csv'
input_file = os.path.join(folderlink+folder_output, ddrc_file)
```

```python
df = pd.read_csv(input_file, sep=';', encoding='utf-8')
```

```python
sparql_endpoint = "https://api.triplydb.com/datasets/academy/pokemon/services/jena/sparql"
sparql_query = """
SELECT ?sub ?pred ?obj WHERE {
    ?sub ?pred ?obj .
}
ORDER BY ?pred
LIMIT 10
"""
```

```python
sparql = SPARQLWrapper(sparql_endpoint)
sparql.setQuery(sparql_query)
sparql.setReturnFormat(JSON)

results = sparql.query().convert()
```

```python
bindings = results["results"]["bindings"]
data = []

for item in bindings:
    row = {}
    for key in item:
        row[key] = item[key]["value"]
    data.append(row)

df = pd.DataFrame(data)
```
